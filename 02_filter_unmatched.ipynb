```python
# import packages
import pandas as pd

# define variables
variable_mapping = {
    '21003-2.0': 'age',
    '31-0.0': 'sex',
    '22189-0.0': 'tdep',
    '20197-2.0': 'pal', #verbal memory 
    '2060-2.0': 'freq_unenthusiasm_disinterest', #anhedonia
    '1031-2.0': 'freq_social_visits' #social functioning
}

internalizing = ['20446','20421','20425','20479','20485','29132','29144']
psychotic      = ['20501','20468','20474','20463','20471']
externalizing  = ['20401']
hitop = internalizing + psychotic + externalizing

cols_to_keep = ['eid'] + list(variable_mapping.keys()) + hitop

# load behavioral data
df = pd.read_csv(
    'current.csv',
    usecols=cols_to_keep,
    index_col='eid',
    dtype=str
)
df.rename(columns=variable_mapping, inplace=True)

# load dx dataset
dx = pd.read_csv('ukbb_dx.csv', index_col='eid', dtype=str)
dx_filtered = dx.filter(like='dx')

# merge datasets, keeping only participants present in dx
filtered_unmatched = df.join(dx_filtered, how='inner')

# save filtered dataset
filtered_unmatched.to_csv('filtered_unmatched.csv')

# check
print(filtered_unmatched.head())
print(f"Number of participants: {len(filtered_unmatched)}")
```
